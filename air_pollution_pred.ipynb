<h2><b>Caraterísitcas de las variables en Dataset</b></h2>
<h4><b>CO (Monóxido de carbono)<b><h4>El monóxido de carbono es un gas incoloro, inodoro e insípido que se produce principalmente por la combustión incompleta de combustibles fósiles como la gasolina, el gas natural, el carbón y la madera. Las fuentes comunes de CO incluyen los vehículos de motor, estufas, hornos y chimeneas. En ambientes exteriores, el CO puede acumularse en áreas con alta densidad de tráfico o en ciudades con poca ventilación atmosférica.<br><br><h4><b>NO (Óxido Nítrico)<b><h4>El óxido nítrico es un gas incoloro formado por la combinación de nitrógeno y oxígeno. Es un precursor en la formación de dióxido de nitrógeno (NO2) y de otros óxidos de nitrógeno (NOX). Aunque es un gas reactivo y de corta vida, juega un papel importante en la química atmosférica y en la formación de contaminantes secundarios como el ozono troposférico. Se emite principalmente en procesos de combustión, como los de vehículos y plantas industriales.<br><br><h4><b>NO<sub>2</sub> (Dióxido de Nitrógeno)<b><h4>El dióxido de nitrógeno es un gas tóxico de color marrón rojizo que resulta de la oxidación del óxido nítrico (NO) en la atmósfera. Es uno de los principales contaminantes del aire y contribuye a la formación de partículas finas (PM) y ozono troposférico. Es emitido principalmente por fuentes de combustión como automóviles, plantas de energía y procesos industriales. La exposición a NO2 está asociada con problemas respiratorios y otras afecciones de salud.<br><br><h4><b>NOX (Óxidos de Nitrógeno)<b><h4>Los óxidos de nitrógeno (NOX) son un grupo de gases compuestos principalmente de óxido nítrico (NO) y dióxido de nitrógeno (NO2). Estos gases son importantes precursores en la formación de ozono troposférico y material particulado. NOX se generan en todas las actividades que involucran la quema de combustibles fósiles, como la generación de energía, el transporte y ciertas industrias. Son un componente clave en la contaminación del aire y tienen efectos significativos sobre la salud y el medio ambiente.<br><br><h4><b>O3 (Ozono Troposférico)<b><h4>El ozono troposférico es un gas incoloro que se forma en la capa inferior de la atmósfera (troposfera) a través de reacciones químicas entre los óxidos de nitrógeno (NOX) y compuestos orgánicos volátiles (COV) en presencia de la luz solar. Aunque el ozono en la estratosfera (capa superior) es beneficioso y protege contra la radiación ultravioleta, el ozono troposférico es un contaminante nocivo que afecta la salud respiratoria y daña cultivos y ecosistemas.<br><br><h4><b>PM10 (Material Particulado de 10 micrómetros o menos)<b><h4>PM10 se refiere a partículas sólidas o líquidas suspendidas en el aire con un diámetro aerodinámico de 10 micrómetros o menos. Estas partículas pueden ser inhaladas y llegar al tracto respiratorio superior, lo que puede provocar efectos adversos en la salud, como irritación de las vías respiratorias, enfermedades cardíacas y pulmonares. Las fuentes comunes de PM10 incluyen el polvo de carreteras, la quema de combustibles fósiles, las actividades industriales y la erosión del suelo.<br><br><h4><b>PM2.5 (Material Particulado de 2.5 micrómetros o menos)<b><h4>PM2.5 consiste en partículas finas con un diámetro aerodinámico de 2.5 micrómetros o menos. Estas partículas son especialmente preocupantes porque pueden penetrar profundamente en los pulmones y entrar en el sistema circulatorio, causando efectos graves en la salud, incluyendo enfermedades respiratorias y cardiovasculares. PM2.5 proviene de diversas fuentes, como la combustión de combustibles fósiles, incendios forestales, y procesos industriales.<br><br><h4><b>PRS (Pressure, Prsión Atmoeférica)<b><h4>La presión atmosférica es la fuerza ejercida por la atmósfera sobre la superficie terrestre, medida en unidades como hectopascales (hPa) o milibares (mb). Es un parámetro meteorológico clave que afecta la circulación del aire y la dispersión de contaminantes en la atmósfera. La presión atmosférica varía con la altitud y las condiciones climáticas, y su monitoreo es esencial para entender la dinámica de los contaminantes del aire.<br><br><h4><b>RH (Relative Humidity, Humedad Relativa)<b><h4>La humedad relativa es la cantidad de vapor de agua presente en el aire expresada como un porcentaje del máximo que podría contener a esa temperatura. La humedad relativa afecta la formación y dispersión de contaminantes atmosféricos, como el ozono y las partículas finas. Niveles altos de humedad relativa pueden favorecer la formación de partículas secundarias y modificar la percepción del calor en el ambiente.<br><br><h4><b>SR (Solar Radiation, Radiación Solar)<b><h4>La radiación solar es la cantidad de energía solar recibida por la superficie terrestre, medida en unidades de radiación como vatios por metro cuadrado (W/m²). La radiación solar es fundamental para la formación de ozono troposférico y otros procesos fotoquímicos que afectan la calidad del aire. Además, influye en la temperatura y otros aspectos del clima que tienen un impacto en la concentración y dispersión de contaminantes.

### El Factor de Inflación de la Varianza (VIF) es una medida que indica cuánto se inflan las varianzas de los coeficientes de regresión debido a la multicolinealidad entre las variables predictoras.

$$VIF(X_i)=\frac{1}{1-R_i^2}$$

donde $R_i^2$ es el coeficiente de determinación de la regresión lineal múltiple

$$X_i = \beta_0+\sum_{j\neq i}\beta_jX_j$$

## Interpretación del VIF
* $VIF = 1$: No hay correlación entre esta variable predictora y las demás variables predictoras.
* $1 < VIF < 5$: Correlación moderada, generalmente no problemática.
* $VIF > 5$: Alta correlación, puede ser problemática.
* $VIF > 10$: Muy alta correlación, se recomienda tomar medidas para reducir la multicolinealidad.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from patsy import dmatrices

# Cálculo del factor de inflación de la varianza
# Renombrando una columna
df_temp = df.rename(columns={'PM2.5': 'PM2_5'})

formula = 'PM2_5 ~ CO + NO + NO2 + NOX + O3 + PM10 + PRS + RH + SR + TOUT'
y, X = dmatrices(formula, df_temp, return_type='dataframe')

vif = pd.DataFrame()
vif['Variable'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
#print(vif)

# Revertir el nombre de la columna
df_temp = df_temp.rename(columns={'PM2_5': 'PM2.5'})
vif.sort_values(by='VIF', ascending=False)
```

<b>Clustering</b>
<h4>Se busaca el Silhouette adecuado para determinar el numero de clusters que se van a usar. Se usará el método K-means ya que este nos va a proporcionar un métod visual de deternimar la cantidad de n-clusters a determinar, basándonos en el que mayor puntaje obtenga. En este caso es 2.

```python
# df = df.reset_index()
```

```python
from sklearn.cluster import KMeans
from sklearn.discriminant_analysis import StandardScaler
from sklearn.metrics import silhouette_score

X = df.drop(columns=['Date', 'NO', 'NO2', 'NOX'])

X_initial = X.copy()
X[X.columns] = StandardScaler().fit_transform(X)

range_clusters = range(2, 11)
silhouette_scores = []
for k in range_clusters:
    kmeans = KMeans(n_clusters=k, random_state=42)
    cluster_labels = kmeans.fit_predict(X)

    score = silhouette_score(X, cluster_labels)
    silhouette_scores.append(score)

best_k = max(enumerate(silhouette_scores, start=2), key=lambda x: x[1])[0]
print(f'El número óptimo de clusters es: {best_k}')

silhouette_df = pd.DataFrame({'n_clusters': range_clusters, 'silhouette_score': silhouette_scores})
px.line(silhouette_df, x='n_clusters', y='silhouette_score', title='Silhouette Score vs Número de clusters', markers=True)
```

<b>Forecasting<b>
<h4>Se realizó un análisis detallado de la serie temporal de PM2.5, descomponiéndola en sus componentes principales y utilizando técnicas de forecasting para prever su comportamiento a futuro.

<h4>Descomposición de la serie temporal.
<h4>Al descomponer la serie temporal de la variable PM2.5 utilizando la función seasonal_decompose. La serie temporal fue descompuesta en tres componentes:

<h4>1. Tendencia: Captura el comportamiento a largo plazo de la serie, enseñando si existe un aumento, decremento o estabilidad en PM2.5 a lo largo del tiempo.

<h4>2. Estacionalidad: Refleja los patrones cíclicos recurrentes en la serie en un periodo fijo, para este caso de 12 meses. La estacionalidad es importante para identificar los fenómenos que ocurren con regularidad.

<h4>3. Residuo: Representa la parte de la serie que no puede ser explicada ni por la tendencia ni por la estacionalidad, es decir, son fluctuaciones aleatorias.

<h4>La descomposición fue realizada en un modelo aditivo. Posteriormente, se utilizó Prophet como método de forecasting, forecasting hace referencia a la predicción de valores futuros basados en su comportamiento histórico, en este caso pronosticar los niveles de PM2.5 para un periodo de 60 meses.
Prophet es una herramienta desarrollada por Facebook, eficiente para modelar series temporales con componentes estacionales y con tendencias a largo plazo


```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose


df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
series = df['PM2.5']
result = seasonal_decompose(series, model='additive', period=12)
plt.figure(figsize=(14, 10))

plt.subplot(4, 1, 1)
plt.plot(series, label='Serie Original')
plt.title('Serie Temporal Original')
plt.legend(loc='upper left')

plt.subplot(4, 1, 2)
plt.plot(result.trend, label='Tendencia')
plt.title('Tendencia')
plt.legend(loc='upper left')

plt.subplot(4, 1, 3)
plt.plot(result.seasonal, label='Estacionalidad')
plt.title('Estacionalidad')
plt.legend(loc='upper left')

plt.subplot(4, 1, 4)
plt.plot(result.resid, label='Residuo')
plt.title('Residuo')
plt.legend(loc='upper left')

plt.tight_layout()
plt.show()
```

```python
from prophet import Prophet

data = df.reset_index()[['Date', 'PM2.5']].rename(columns={'Date': 'ds', 'PM2.5': 'y'})

model = Prophet()
model.fit(data)

future = model.make_future_dataframe(periods=60, freq='M', include_history=False)
forecast = model.predict(future)
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=data['ds'], y=data['y'], mode='lines', name='Real'))
fig.add_trace(go.Scatter(x=forecast['ds'], y=forecast['yhat'], mode='lines', name='Predicción', line_color='green'))
fig.add_trace(go.Scatter(x=forecast['ds'], y=forecast['yhat_lower'], mode='lines', line_color='green'))
fig.add_trace(go.Scatter(x=forecast['ds'], y=forecast['yhat_upper'], mode='lines', line_color='green', fill='tonexty', fillcolor='rgba(0, 128, 0, 0.45)'))

fig.update_layout(title_text='Pronóstico de PM2.5 para 2016-2020')
fig.show()
```
